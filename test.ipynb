# pembuatan diagram untuk visualisasi

### 1. diagram bar

```python
data = pd.DataFrame({
    "A": [10,30,50,70,90],
    "B": [100,200,300,400,500]
})

data
```

```python
# pembuatan chart dengan bantuan matplotlib kita bs menggunakan
test = plt.bar(data["A"], data["B"], width=12) # uruanya adalah d=sumbu X lalu sumbu Y
plt.title("Contoh penggunaan Bar") # Penggunaan untuk judul
plt.xlabel("Label untuk data A") # pengguanaan untuk penamaan label X
plt.ylabel("Label untuk data b") # Penggunaan untuk penamaan label Y

plt.bar_label(test, label_type="center") # untuk melihat value masing masing bar
plt.show()
```

```python
## atau dengan kita bs membuat langsunng di dataFrame

test = data.plot(kind="bar",x="A", y="B")
plt.title("Contoh penggunaan Bar") # Penggunaan untuk judul
plt.xlabel("Label untuk data A") # pengguanaan untuk penamaan label X
plt.ylabel("Label untuk data b") # Penggunaan untuk penamaan label Y
plt.show()
```

### 3. pie chart

```python
data = pd.DataFrame({
    "labels": ["Python", "Js", "Java", "C++"],
    "minat": [120, 300, 50, 25],
    "explode": [0, 0, 0,0.1]
    
})
plt.pie(x=data.minat, labels=data.labels, center=[0,0], autopct="%1.f%%", explode=data["explode"], shadow=True)
plt.title("Minat dalam bahasa Pemrograman")
plt.show()
```

```python
data.set_index("labels").plot(kind="pie", y="minat", autopct="%1.f%%")
plt.ylabel("")
plt.show()
```

# membuat satacked Bar

```python
categories = ("A", "B", "C", "D")
values_categories = {
    "values1":[10,20,30,40],
    "values2":[15,25,35,45]
}
width= 0.4
bottom = np.zeros(4)

fig, ax =  plt.subplots(figsize = (8,6))

for index, values in values_categories.items():
    bar = ax.bar(categories, values, width, label=index, bottom=bottom)
    bottom+=values ## membuat data / bar dapat di timpa 

    ax.bar_label(bar,label_type="center") # menampilkan nilai di masing masing bar

plt.legend()
```
